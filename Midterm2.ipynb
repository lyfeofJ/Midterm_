```python
import pandas as pd
df = pd.read_csv('amazon-data-2.csv')
df.head()
from matplotlib import pyplot as plt

df = df.fillna(0)
df.head()
df["Item Total"] = df["Item Total"].str.replace('$','').astype(float)
df.head()
df["Item Subtotal"] = df["Item Subtotal"].str.replace('$','').astype(float)
df["Item Subtotal Tax"] = df["Item Subtotal Tax"].str.replace('$','').astype(float)

total = df["Item Total"].sum()
print('Monthly Summary: ', total)
mean = df["Item Total"].mean()
print('Monthly Average: ',mean)
median =  df["Item Total"].median()
print('Median Amount Spent: ',median)
highest_order =  df["Item Total"].max()
print('Most Expensive Order: ',highest_order)
lowest_order =  df["Item Total"].min()
print('Least Expensive Order: ',lowest_order)
standard_deviation = df["Item Total"].std()
print('Standard Deviation: ',standard_deviation)
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()

bar = df.plot.bar(x = 'Order Date', y = 'Item Total', rot=90, figsize=(20,10))
bar.set_xlabel("Order Date")
bar.set_ylabel("$ Spent")

pie = df.plot(kind='pie', y = 'Item Total',figsize=(12,12), labels = None, radius = 1.2)
#plt.legend(loc="center right", fontsize= 10)


plt.show()
```
